```python
import pandas as pd
import numpy as np
das = pd.read_csv('billboard_data.csv')
das
```

```python
l = np.asarray(das[['Title','Artist(s)','year']]).tolist()
```

```python
import re
from collections import defaultdict 
dic = defaultdict(list)
for e in l:
    song_name= ''
    if  '\"This Old Heart of Mine' in e:
        song_name = 'This Old Heart of Mine'
    else:    
        song_name =  re.findall(r'"(.*?)"',e[0])
    #print(song_name)
    if 'featuring' in e[1]:
        art = e[1].split('featuring')[0]
        dic[art].append((song_name, e[2]))
    elif '&' in e[1]:
        art = e[1].split('&')[0]
        dic[art].append((song_name,e[2]))
    else:
        dic[e[1].replace('\n','')].append((song_name,e[2]))
        
dic.pop('Artist(s)')
```

```python
dic2 = defaultdict(list)

for art in dic.keys():
    artist_name = art
    for track in dic[art]:
        track_name = track[0][0]
        track_year = track[1]
        
        results = sp.search(q=track_name+' '+artist_name, limit=1)
        for i, t in enumerate(results['tracks']['items']):
            #print(' ', i, t['name'], t['id'])
            data5 = rq.get("https://api.spotify.com/v1/audio-features/"+t['id'], headers = {'Authorization': "Bearer {}".format(token)}).json()
            track_id = t['id']
            dic2[track_id].append(track_name)
            dic2[track_id].append(artist_name)
            dic2[track_id].append(track_year)
            dic2[track_id].append(data5)

```

```python
analysis_df = pd.DataFrame(dic2.items())
```
